```python
import urllib.request as req
import pandas as pd #for show histogram
import matplotlib.pyplot as plt
import random

CSV_URL = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.csv"

data = req.urlopen(CSV_URL)
names = ["num_pregnancies", "glucose_2hr", "diastolic_bp", "triceps_skinfold", "insulin_2hr" ,"bmi" ,"diabetes_pedigree" ,"age" ,"diabetes_res"]
df = pd.read_csv(data, header=None, names=names)

new_df = df.drop(columns="diabetes_res")

for col in new_df.columns:
  r = random.random()
  g = random.random()
  b = random.random()
  color = (r, g, b)
  plt.hist(df[col], color=color)
  plt.title(col)
  plt.show()
```

```python
datasets = {}

dfByClass =df[["glucose_2hr", "diabetes_res"]].groupby("diabetes_res")

for group, data in dfByClass:
  datasets[group] = data

plt.hist(datasets[0]["glucose_2hr"], label="class 0", density=True, alpha=0.9)
plt.hist(datasets[1]["glucose_2hr"], label="class 1", density=True, alpha=0.9)
plt.legend()
plt.show()
```
