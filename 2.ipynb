```python
import pandas as pd
import urllib.request as request
import matplotlib.pyplot as plt

csv = request.urlopen("https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv")
df = pd.read_csv(csv)

for group, data in df.groupby("species"):
    plt.scatter(data["sepal_length"], data["sepal_width"], label=group)

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Iris Sepal Length by Species")
plt.legend()
plt.show()
```
